```python
# importing dependencies
import pandas as pd
import math
```

```python
# reading in summary csv
summary_df = pd.read_csv('Resources/cluster_summary.csv')
summary_df
```

```python
# calculating total annual kwh
annual_sum_kwh = summary_df['kwh_ann_tot'].sum()
annual_sum_kwh
```

```python
# calculating building specific energy usage
building_kwh = summary_df[['building_type', 'kwh_ann_tot']].groupby('building_type').sum()

# sorting dataframe
building_kwh = building_kwh.sort_values('kwh_ann_tot', ascending=False)

building_kwh
```

```python
# creating list of building types of concern (excluding other)
building_list = summary_df['building_type'].unique()[1:]

# creating cluster list from building types to look up
cluster_list = summary_df.loc[summary_df['building_type'].isin(building_list)]['cluster']

# creating empty dataframe to append to
master_df = pd.DataFrame()

# looping through cluster list to open CSVs and append to master_df
for cluster in cluster_list:
    
    csv_item = "Resources/" + cluster + ".csv"
    
    building = cluster.split('-')[3]
    
    ds = pd.read_csv(csv_item)
    ds['building_type'] = building
    
    master_df = master_df.append(ds)
    
master_df
```

```python
# calculating kwh per hour 
kwh_hr_df = master_df[['hour_ending', 'total']].groupby('hour_ending').sum()

# sorting list
kwh_hr_df = kwh_hr_df.sort_values('total', ascending=False)

kwh_hr_df
```

```python
# grabbing peak hour
peak_hr = int(kwh_hr_df.index[0])

# calculating end user contribution
end_use_peak = master_df.groupby('hour_ending').sum().iloc[peak_hr]

# sorting values
end_use_peak = end_use_peak.sort_values(ascending=False)

end_use_peak
```

```python
# creating dataframe of hourly usage by building type - resetting index due to hour_ending conflicts
peak_building_df = master_df.groupby(['hour_ending', 'building_type']).sum().reset_index()

# grabbing peak hour from peak building df
peak_building = peak_building_df[['hour_ending', 'building_type', 'total']].loc[peak_building_df['hour_ending'] == peak_hr]

# sorting df
peak_building =  peak_building.sort_values('total', ascending=False)

peak_building
```

```python
# determining start and end hours for the peak day
start_hr = math.floor(365 - ((8760 - peak_hr) / 24)) * 24
end_hr = math.ceil(365 - ((8760 - peak_hr) / 24)) * 24

# creating plot for peak day's usage
master_df.loc[(master_df['hour_ending'] >= start_hr) & (master_df['hour_ending'] < end_hr)].groupby('hour_ending').sum().plot(
    kind='bar',
    figsize=(14,8)
)
```

```python
# creating "hour of day" for each entry
master_df['day_hour'] = master_df['hour_ending'] % 24

# grouping by hour of day and averaging - note used this method instead of .mean() due to some items not have values.. they're different
avg_hr_df = master_df.groupby('day_hour').sum() / 365

# dropping hour ending for plotting
avg_hr_df = avg_hr_df.drop(columns = {'hour_ending'})

# plotting data
avg_hr_df.plot(
    kind='bar',
    figsize=(14,8)
)
```

```python
# relatively 'same' as sum std? (what a cluster..)
master_df.groupby('day_hour').std().sum()
```

```python
# list of buildings of concern
bld_list = ['office', 'retail']

# creatning dataframe of data based on buildings in bld list
building_lighting_df = master_df[['building_type', 'lighting', 'total']].loc[master_df['building_type'].isin(bld_list)].groupby('building_type').sum()

# calcualted theoretical efficiency ratio
efficiency = 80/110

# calculating LED usage column based on eff ratio
building_lighting_df['lighting_led'] = building_lighting_df['lighting'] * efficiency

# calcualting led savings
building_lighting_df['led_savings'] = building_lighting_df['lighting'] - building_lighting_df['lighting_led']

# calculating total percent savings
building_lighting_df['total_pct_savings'] = (building_lighting_df['led_savings'] / building_lighting_df['total'].sum()) * 100

building_lighting_df
```
